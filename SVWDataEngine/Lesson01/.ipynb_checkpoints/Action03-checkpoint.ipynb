```python
import numpy as np
import pandas as pd

df = pd.read_csv(r"car_complain.csv",encoding='utf-8')
df.head()
```

```python
res = df.join(df['problem'].str.get_dummies(','))
res.head()
```

```python
problem_bybrandmodel_counts = res.groupby(['brand','car_model'])['id'].agg(['count'])
print(problem_bybrandmodel_counts)
```

```python
problem_bybrand_counts = res.groupby(['brand'])['id'].agg(['count'])
print(problem_bybrand_counts)
```

```python
tags = res.columns[7:]
tags
```

```python
df2= res.groupby(['brand'])[tags].agg(['sum'])
df2
```

```python
df= res.groupby(['brand'])['id'].agg(['count'])
df
```

```python
df2 = df.merge(df2, left_index=True, right_index=True, how='left')
df2
```

```python
df2.reset_index(inplace=True)
df2
```

```python
df2= df2.sort_values('count', ascending=False)
df2
```

```python
query = ('A11', 'sum')
print(df2.sort_values(query, ascending=False))
```
